# Bài Tập Về Nhà 01 :
Tính đạo hàm của hàm số y = 3x^5 + 2x^4 + x^3 + 4x + 7 tại x = 3 sử dụng PyTorch
Cho x = 5x^5 + 6x^3 -3x +1 
Cho biết độ dốc của đa thức trên ở điểm nào 

```python
import torch

# y = 3x^5 + 2x^4 + x^3 + 4x + 7 tại x=3
x1 = torch.tensor(3.0, requires_grad=True)
y1 = 3*x1**5 + 2*x1**4 + x1**3 + 4*x1 + 7
y1.backward()
print("y'(3) =", x1.grad.item())  
# y = 5x^5 + 6x^3 - 3x + 1: điểm có độ dốc = 0
def slope(x_val):
    x = torch.tensor(float(x_val), requires_grad=True)
    y = 5*x**5 + 6*x**3 - 3*x + 1
    y.backward()
    return x.grad.item()
print("Điểm dốc=0: x≈+0.3736, y' =", slope(0.3736))
print("Điểm dốc=0: x≈-0.3736, y' =", slope(-0.3736))
```

# Bài Tập Về Nhà 02 :
Tạo một tensor ban đầu có giá trị là 2 
Định nghĩa hàm số và tính gradient 
y = x^3 + 2x^2 + 5x + 1
hãy tính dy/dx tại giá trị của x 
Dùng phương pháp Gradient Desccent với Learning rate alpha = 0.1 để cập nhật giá trị x trong 10 vòng lập

```python
import torch

alpha = 0.1
so_vong = 10
x = torch.tensor(2.0, requires_grad=True)
y = x**3 + 2*x**2 + 5*x + 1
y.backward()
print("Ban đầu: x =", x.item(), "| dy/dx =", x.grad.item())
for i in range(1, so_vong + 1):
    y = x**3 + 2*x**2 + 5*x + 1
    y.backward()
    grad = x.grad.item()
    with torch.no_grad():
        x -= alpha * x.grad
    x.grad.zero_()
    print(f"Vòng {i:02d}: x = {x.item():.6f} | dy/dx = {grad:.6f}")
```

# Bài Tập Về Nhà 03 :
Tạo một tập dữ liệu giả lập với x là số giờ học ( ngẫu nhiên từ 1 - 10 ) và y là số điểm được tính theo công thức y  =3x + 5 + noise
Với noise là một giá trị ngẫu nhiên nhỏ 

1. Khởi tạo tham số w va b nhẫu nhiên với requires_grad = True
2. Tính MSE
3. Tính Gradient 
4. Cập nhật tham số w và b bằng Gradient Desccent với Learning rate alpha = 0.01
5. Lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của mô hình .

```python
import torch
torch.manual_seed(42)
n = 50
x = torch.randint(1, 11, (n, 1), dtype=torch.float32)   # giờ học 1..10
noise = torch.randn(n, 1) * 0.5                          # nhiễu nhỏ
y = 3 * x + 5 + noise

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
alpha = 0.01

for epoch in range(1, 101):
    y_hat = w * x + b                    # dự đoán
    loss = ((y_hat - y) ** 2).mean()     # 2) MSE
    loss.backward()                      # 3) Gradient
    with torch.no_grad():                # 4) Update
        w -= alpha * w.grad
        b -= alpha * b.grad
    w.grad.zero_()                       # reset grad
    b.grad.zero_()

    if epoch % 10 == 0 or epoch == 1:    # 5) xem hội tụ
        print(f"Epoch {epoch:03d} | MSE={loss.item():.4f} | w={w.item():.4f} | b={b.item():.4f}")

print("\nKết quả cuối: w ~", w.item(), " | b ~", b.item())
```

# Bài Tập Về Nhà 05 : 
Bạn giúp thầy tạo tensor với
1. empty 
2. zeros
3. ones
4. random
5. Reshape với view và view as

```python
import torch

# Công nghệ: PyTorch (torch) - Tensor API + reshape (view, view_as)

# 1) empty
a = torch.empty(2, 3)
print("empty:\n", a)

# 2) zeros
b = torch.zeros(2, 3)
print("\nzeros:\n", b)

# 3) ones
c = torch.ones(2, 3)
print("\nones:\n", c)

# 4) random
d = torch.rand(2, 3)
print("\nrandom:\n", d)

# 5) reshape với view và view_as
e = torch.arange(1, 13)      # 1..12
e_view = e.view(3, 4)        # view
mau = torch.zeros(2, 6)
e_view_as = e.view_as(mau)   # view_as

print("\nview(3,4):\n", e_view)
print("\nview_as(2,6):\n", e_view_as)
```
